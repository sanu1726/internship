```python
data.target.value_counts()
```

```python
data.target.value_counts().plot(kind = 'bar', color=["red", "blue"])
plt.title("Heart Disease values")
plt.xlabel("0 = No  Heart Disease, 1 = Heart Disease")
plt.ylabel("Amount")

```

```python
data.target.value_counts().plot(kind = 'pie', figsize= (8,6))
plt.legend(["No Heart Disease", "Heart Disease"])
plt.title("Heart Disease values")
```

```python
data.sex.value_counts()
```

```python
data.sex.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.title("MALE FEMALE Ratio")
plt.legend(["Male", "Female"])
```

```python
pd.crosstab(data.sex, data.target)
```

```python
sns.countplot(x='target', data = data, hue = 'sex')
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("0 = No Heart Disease, 1 = Heart Disease")
```

```python
data.cp.value_counts()
```

```python
data.cp.value_counts().plot(kind = 'bar', color = ['salmon', 'lightskyblue', 'springgreen', 'khaki'])
plt.title('Chest pain type vs count');
```

```python
pd.crosstab(data.sex, data.cp)
```

```python
pd.crosstab(data.sex, data.cp).plot(kind = 'bar', color =['coral', 'lightskyblue','plum','khaki'])
plt.title("This of Chest pain for Sex")
plt.xlabel('0=Female. 1=Male');
```
